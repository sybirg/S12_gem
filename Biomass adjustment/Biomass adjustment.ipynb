```python
# DNA and AA
# call the sequence file
# protein sequence or dna sequence

f = open('~\S12 protein sequence', 'r')

lines = f.readlines()

new = ''

for i, l in enumerate(lines):
    if '[pseudo=true]' not in l: # not pseudo gene
        if l.startswith('>') and '[pseudo=true]' not in l: # search the locus tag
            continue
        else: # add the sequence
            new+=l

f.close()
```

```python
# counting

count = {}
for i in new:
    if i == '\n':
        continue
    else:
        try:
            count[i] +=1
        except:
            count[i] = 1
            
count # apply the result
```

```python
# RNA
# call the file

import pandas as pd

pd.set_option('display.max_colwidth', None)

df = pd.read_excel("S12.xlsx", engine = "openpyxl")
```

```python
def dna_to_rna(dna):
    change_to = {'G':'C','C':'G','T':'A','A':'U'}
    rna = ''.join([change_to[i] for i in dna])
    return rna
```

```python
mRNA = dna_to_rna(df[df['Type']=='G']['Seqn'].sum())
tRNA = dna_to_rna(df[df['Type']=='T']['Seqn'].sum())
rRNA = dna_to_rna(df[df['Type']=='R']['Seqn'].sum())
```

```python
atp_rna = df[df['metabolite']=='atp**']['mmol/gDW'].item()
pi_inorg = df[df['metabolite']=='pi**']['mmol/gDW'].item()
h2o_aa = round(df.iloc[:20]['mmol/gDW'].sum(),4)
```

```python
# q= growth_rate/Y_G + maintenance coefficient(mc)

inverse_Y_G = 10
mc = 0.077



# NGAM calculation

import numpy as np

gam_am = {'h2o_c','atp_c','adp_c','h_c','pi_c'}

for m in gam_am:
    globals()[m] = model.metabolites.get_by_id(m)
    globals()[m+'_coef'] = biomass_eq.get_coefficient(globals()[m])

with model:
    model.reactions.get_by_id('ATPM').bounds = (-99999,99999)
    biomass_eq.add_metabolites({'h2o_c':-h2o_c_coef,'atp_c':-atp_c_coef,'adp_c':-adp_c_coef,'h_c':-h_c_coef,'pi_c':-pi_c_coef})

    model.objective = 'ATPM'
    model.reactions.get_by_id("EX_glc__D_e").lower_bound = -mc
    biomass_eq.lower_bound = 0
    ngam = round(model.optimize().objective_value,2)
        
print('NGAM:',ngam)
```

```python
def nearest_value(simulated_data, inverse_Y_G):
    
    tmp = 0
    gam = 0
    
    for i in simulated_data.items():
        if abs(inverse_Y_G-tmp) > abs(inverse_Y_G-i[1]):
            gam = i[0]
            tmp = i[1]
    
    return gam
```

```python
# GAM calculation

model.reactions.get_by_id('ATPM').bounds = (ngam,ngam) # fix the ngam

gam_am = {'h2o_c','atp_c','adp_c','h_c','pi_c'}

glucose = [1,2,3,4,5] # mmol

simulated_data = {}

for coef in (np.arange(42.30,42.35,0.01)): # you can change the range
    objective_values = []
    
    with model:
        for glc in glucose:
            model.reactions.get_by_id("EX_glc__D_e").lower_bound = -glc
            
            for m in gam_am:
                globals()[m+'_coef'] = biomass_eq.get_coefficient(globals()[m])
                
            biomass_eq.add_metabolites({'h2o_c':-h2o_c_coef,'atp_c':-atp_c_coef,'adp_c':-adp_c_coef,'h_c':-h_c_coef,'pi_c':-pi_c_coef})
            
            biomass_eq.add_metabolites({'h2o_c':round(-(coef-h2o_aa),6),'atp_c':round(-(coef+atp_rna),6),'adp_c':round(coef,6),'h_c':round(coef,6),'pi_c':round(coef-pi_inorg,6)})

            objective_values.append(model.optimize().objective_value)
            
        slope, intercept = np.polyfit(objective_values, glucose,1)
        simulated_data[coef]=slope
        print(coef, slope)

        
# get the value which most nearest with inverse_Y_G

gam = round(nearest_value(simulated_data, inverse_Y_G),2)
print('GAM:',gam)
```

```python
#fix the GAM/NGAM

gam_am = {'h2o_c','atp_c','adp_c','h_c','pi_c'}

for m in gam_am:
    globals()[m] = model.metabolites.get_by_id(m)
    globals()[m+'_coef'] = biomass_eq.get_coefficient(globals()[m])

biomass_eq.add_metabolites({'h2o_c':-h2o_c_coef,'atp_c':-atp_c_coef,
                            'adp_c':-adp_c_coef,'h_c':-h_c_coef,'pi_c':-pi_c_coef})

biomass_eq.add_metabolites({'h2o_c':round(-(gam-h2o_aa),6),'atp_c':round(-(gam+atp_rna),6),
                            'adp_c':round(gam,6),'h_c':round(gam,6),'pi_c':round(gam-pi_inorg,6)})

print(biomass_eq.reaction)
    
model.reactions.get_by_id('ATPM').bounds = (ngam, 999999.0)
```
